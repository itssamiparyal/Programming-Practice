```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df
```

```python
df.info()
```

```python
"Missing Values:", df.isnull().sum()
```

```python
df = df.dropna(subset=['Year','Publisher'])
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
"Missing Values:", df.isnull().sum()
```

# Platform Distribution (Bar Chart)

```python
df['Platform'].value_counts().plot(kind='bar', color='purple')
plt.title("Game Count by Platform")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.show()
```

# Platform Distribution (Pie Chart)

```python
df['Platform'].value_counts().plot(kind='pie')
plt.title("Game Distribution by Platform")
plt.ylabel("")
plt.show()
```

# Yearly Game Releases (Line Chart)

```python
df['Year'].value_counts().sort_index().plot(kind='line', color='blue', marker='o')
plt.title("Games Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```

# Global Sales Histogram

```python
df['Global_Sales'].hist(color='teal')
plt.title("Distribution of Global Sales")
plt.xlabel("Global Sales (millions)")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Genre'].value_counts().plot(kind='pie')
plt.title("Game Distribution by Genre")
plt.ylabel("")
plt.show()
```

```python
print("Conclusion:")
print("1. The number of games released per year shows trends over time. This  indicating growth or decline in game development.")
print("2. The distribution of global sales Says  that most games sell in smaller quantities.While only a few achieving massive sales.")
print("3. The pie chart shows how games are distributed across different platforms. It highlightes dominant ones.")
print("4. The bar chart provides a clearer view of the most popular platforms based on the number of games released.")
```
