```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
male_count = (df.Sex == 'male').sum()
```

```python
female_count =(df.Sex == 'female').sum()
```

```python
plt.bar(['Male','Female'],[male_count,female_count])
plt.xlabel('Gender')
plt.ylabel('Number')
plt.title('Gender count')
plt.show()
```

```python
plt.pie([male_count,female_count],labels=['Male','Female'],autopct ='%.2f%%')
plt.show()
```

```python
plt.hist(df.Age, bins=15)
plt.title('Age Interval')
plt.xlabel('Interval')
plt.ylabel('Age')
plt.show()
```
