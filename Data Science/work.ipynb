```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
steam_games_path = "steam_games.csv" 
df = pd.read_csv(steam_games_path)
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df["Year"] = df["release_date"].dt.year
```

```python
df = df.dropna(subset=["Year"])
df["Year"] = df["Year"].astype(int)
```

```python
platform_counts = df["platforms"].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 5))
platform_counts.plot(kind="bar", color="purple")
plt.title("Game Count by Platform")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
platform_counts.plot(kind="pie")
plt.title("Game Distribution by Platform")
plt.ylabel("")
plt.show()
```

```python
yearly_counts = df["Year"].value_counts().sort_index()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(yearly_counts.index, yearly_counts.values, marker="o", linestyle="-", color="blue")
plt.title("Games Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df["price_initial (USD)"].hist(bins=30, color="green", edgecolor="black")
plt.title("Distribution of Game Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Number of Games")
plt.show()
```

```python
genre_counts = df["genres"].value_counts().head(10)
```

```python
plt.figure(figsize=(8, 8))
genre_counts.plot(kind="pie")
plt.title("Game Distribution by Genre")
plt.ylabel("")
plt.show()
```
